```python
# to remove /n from above prices
import re

def remove(strings):
  pt = re.compile(r"[\n]")
  modified_strings = []
  for string in strings:
    modified_string = re.sub(pt, "", string)
    modified_strings.append(modified_string)
  return modified_strings


result = remove(price)
print(result)
```
