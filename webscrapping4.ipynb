```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
name=[]
artist=[]
date=[]
views=[]
```

```python
tag2 = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
for i in tag2:
    try:
        nam= i.text
        name.append(nam)
    except NoSuchElementException:
        name.append('-')
        
tag3 = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
for j in tag3:
    try:
        art= j.text
        artist.append(art)
    except NoSuchElementException:
        artist.append('-')
        
tag4 = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
for k in tag4:
    try:
        dat= k.text
        date.append(dat)
    except NoSuchElementException:
        date.append('-')

tag5 = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
for l in tag5:
    try:
        view= l.text
        views.append(view)
    except NoSuchElementException:
        views.append('-')
        
```

```python
import pandas as pd
df=pd.DataFrame({'Video_Name':name,'Artist':artist,'Date_Of_Publish':views,'views(In billion)':date})
df
```
