```python
url= []
for i in range(3):
    link=driver.find_elements(By.XPATH, '//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for j in link:
        url.append(j.get_attribute('href'))
    
    
    
url
        
  
```

```python
brand=[]
name=[]
price=[]
Return=[]
delivery=[]
available=[]

for i in url:
    driver.get(i)
    time.sleep(5)
    try:
        bran=driver.find_element(By.XPATH, '/html/body/div[2]/div/div[5]/div[3]/div[4]/div[47]/div/table/tbody/tr[1]/td[2]/span')
        brand.append(bran.text)
    except NoSuchElementException:
        brand.append('-')
    try:
        nam=driver.find_element(By.XPATH, '/html/body/div[2]/div/div[5]/div[3]/div[4]/div[1]/div/h1/span')
        name.append(nam.text)
    except NoSuchElementException:
        name.append('-')
    try:
        pric=driver.find_element(By.XPATH, '/html/body/div[2]/div/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[3]/div/div[1]/div/div/span[1]/span[2]/span[2]')
        price.append(pric.text)
    except NoSuchElementException:
        price.append('-') 
    try:
        Retur=driver.find_element(By.XPATH, '/html/body/div[2]/div/div[5]/div[3]/div[4]/div[24]/div[2]/div/div/div/div[2]/div/ol/li[3]/div/span/div[2]/span')
        Return.append(Retur.text)
    except NoSuchElementException:
        Return.append('-')
    try:
        deliver=driver.find_element(By.XPATH, '/html/body/div[2]/div/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[4]/div/div[3]/div[10]/div[1]/div/div/div/span/span[1]')
        delivery.append(deliver.text)
    except NoSuchElementException:
        delivery.append('-')
    try:
        availabl=driver.find_element(By.XPATH, '/html/body/div[2]/div/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[4]/div/div[5]/div/div[1]/span')
        available.append(availabl.text)
    except NoSuchElementException:
        available.append('-')
```

```python
import pandas as pd
df= pd.DataFrame({'Name':name, 'Price':price, 'Return Policy':Return,'Delivery by':delivery,'Availability':available})
df
```

```python
name=[]
price=[]
rom=[]
camera=[]
display=[]
battery=[]

for i in url:
    driver.get(i)
    time.sleep(5)

    try:
        nam=driver.find_element(By.XPATH, '//html/body/div[1]/div/div[3]/div[1]/div[2]/div[2]/div/div[1]/h1/span')
        name.append(nam.text)
    except NoSuchElementException:
        name.append('-')
    try:
        pric=driver.find_element(By.XPATH, '/html/body/div[1]/div/div[3]/div[1]/div[2]/div[2]/div/div[4]/div[1]/div/div[1]')
        price.append(pric.text)
    except NoSuchElementException:
        price.append('-') 
    try:
        color=driver.find_element(By.XPATH, '/html/body/div[1]/div/div[3]/div[1]/div[2]/div[6]/div/div[1]/div[1]/div/ul/li[2]/div/div')
        colour.append(color.text)
    except NoSuchElementException:
        colour.append('-')
    try:
        ra=driver.find_element(By.XPATH, '/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[1]')
        rom.append(ra.text)
    except NoSuchElementException:
        rom.append('-')

    try:
        primary=driver.find_element(By.XPATH, '/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[3]')
        camera.append(primary.text)
    except NoSuchElementException:
        camera.append('-')

    try:
        displa=driver.find_element(By.XPATH, '/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[2]')
        display.append(displa.text)
    except NoSuchElementException:
        display.append('-')
    try:
        batter=driver.find_element(By.XPATH, '/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[4]')
        battery.append(batter.text)
    except NoSuchElementException:
        battery.append('-')
```

```python
import pandas as pd
df= pd.DataFrame({'Brand,Name&Colour':name,'Display-SIze':display,'Battery-Capacity':battery,'Ram/Rom':rom,'Camera specs':camera,'Price':price,'Product-URL':url})
df
```
